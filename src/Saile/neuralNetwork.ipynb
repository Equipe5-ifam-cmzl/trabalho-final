```python
import pandas as pd

path = "C:/Users/Saile/Desktop/trabalho ML/Sprint2/"

df_finance = pd.read_csv(path + "Financial-Data Tratado.csv")

df_finance.head()
```

```python
df_finance.corr()
```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression

y = df_finance['e_signed']

X = df_finance[['risk_score', 'amount_requested', 'has_debt', 'income']]

X_train, X_test, y_train, y_test = tts(X, y, test_size=0.3, random_state=42)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras.callbacks import EarlyStopping, ReduceLROnPlateau

redeneural = Sequential()

redeneural.add(Dense(128, activation = 'relu', input_dim = 4))
redeneural.add(Dense(128, activation = 'relu'))
redeneural.add(Dense(128, activation = 'relu'))

redeneural.add(Flatten())
redeneural.add(Dense(1, activation = 'sigmoid'))
```

```python
earlystop = EarlyStopping(monitor='loss', patience = 10)
learning_rate_reduction = ReduceLROnPlateau(monitor = 'accuracy',patience = 2,verbose = 1,factor = 0.5,min_lr = 0.00001)
callbacks = [earlystop,learning_rate_reduction]
```

```python
redeneural.compile(
    loss="binary_crossentropy",
    optimizer="adam",
    metrics=['accuracy']
)
```

```python
redeneural.fit(
    X_train,
    y_train.values,
    epochs=30,
    batch_size=64,
    callbacks = callbacks
)
```

```python
pred = redeneural.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

print(f"MSE : {mean_squared_error(y_test, pred)}")
print(f"MAE : {mean_absolute_error(y_test, pred)}")
```
