```python
import numpy as np
import pandas as pd
from category_encoders import OneHotEncoder
```

```python
# 1. para carregar o dataset

link = 'C:/Users/CarlosViniMSouza/Documents/Projects/data-mining-ifam/database/Financial-Data.csv'

dfMerge = pd.read_csv(link)
dfMerge
```

```python
# 2. Vamos eliminar as colunas com valores muito próximos de 0:
#     ° E para isso, aplicaremos o método de 'Valor Absoluto'
#     ° E a ordenação do dataset usando o 'e-signed' como ref.

# abs(dfMerge.corr()[['e_signed']].sort_values(['e_signed'])['e_signed']) < 0.01

dfMergeModule = abs(dfMerge.corr()[['e_signed']])

dfMergeModuleSorted = dfMergeModule.sort_values(['e_signed'])['e_signed']
dfMergeModuleSorted < 0.01
```

```python
# 3. Todas as colunas que retornaram 'True' na cédula acima serão dropadas

dfMerge = dfMerge.drop([
    'risk_score_5', 'Entry_id', 
    'years_employed', 'risk_score_4'
], axis=1)

dfMerge
```

```python
"""
OBS.:
    1. A codificação OneHot(fictícia) para recursos categóricos produz um recurso por categoria, cada binário.
       (fonte: https://contrib.scikit-learn.org/category_encoders/onehot.html)
    
    2. A coluna 'pay_schedule' possui 4 valores distintos:
       ° 'bi-weekly' | 'weekly' | 'semi-monthly' | 'monthly'
       
    3. Com isso, teremos 4 colunas binárias de 'pay_schedule'
"""

# 4. Vamos desenvolver mais a coluna 'pay_schedule'

one = OneHotEncoder(cols=['pay_schedule']).fit_transform(dfMerge['pay_schedule'])

dfMerge = pd.concat([one, dfMerge.drop(['pay_schedule'], axis=1)], axis=1)
dfMerge
```

```python
# 5. Agora, vamos fazer algumas mudanças em nas colunas de 'personal_account'

# 5.1. Equacionando: P.A. (anual) = (P.A. (ano) * 12) + P.A. (mes)
dfMerge["personal_accounts_month"] = dfMerge["personal_account_y"] * 12 + dfMerge["personal_account_m"]

# 5.2. Feitos as mudanças, podemos remover 'personal_account_y/m'
dfMerge = dfMerge.drop(["personal_account_y", "personal_account_m"], axis=1)

dfMerge
```

```python
# 6. Outra operação que preciso entender ...Sei que envolve logaritmo!

dfMerge['risk_score'] = round(np.log2(dfMerge['risk_score']), 4)
dfMerge
```

```python
# 7. Calculando a '%' de 'renda' & 'qtd. solicitada'

dfMerge['income'] = dfMerge['income'] / 100
dfMerge['amount_requested'] = dfMerge['amount_requested'] / 10
```

```python
# 8. Removendo outras 2 colunas não relevantes

dfMerge = dfMerge.drop(['months_employed', 'age'], axis=1)
dfMerge
```

```python
# 9. Outra operação para entender ...

dfMerge['risk_score_2'] = dfMerge['risk_score_2'] * 10
dfMerge['risk_score_3'] = dfMerge['risk_score_3'] * 10

dfMerge['ext_quality_score'] = dfMerge['ext_quality_score'] * 10
dfMerge['ext_quality_score_2'] = dfMerge['ext_quality_score_2'] * 10

dfMerge
```
